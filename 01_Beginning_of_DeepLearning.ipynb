```python
import numpy as np
import pandas as pd

raw_link = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/x09.txt'
raw_data = np.genfromtxt(raw_link, skip_header=36)
raw_data
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline

xs = np.array(raw_data[:,2], dtype=np.float32)
ys = np.array(raw_data[:,3], dtype=np.float32)
zs = np.array(raw_data[:,4], dtype=np.float32)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs)
ax.set_xlabel('Weight')
ax.set_ylabel('Age')
ax.set_zlabel('Blood fat')
ax.view_init(15, 15)
plt.show()
```

##### 학습 대상 데이터를 추린다.

```python
x_data = np.array(raw_data[:,2:4], dtype=np.float32)
y_data = np.array(raw_data[:,4], dtype=np.float32)
y_data = y_data.reshape((25, 1))
```

```python
# pip install tensorflow
```

```python
import tensorflow as tf
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(1, input_shape=(2, ))
])
model.compile(optimizer = 'rmsprop', loss='mse')
```

```python
model.summary()
```

```python
hist = model.fit(x_data, y_data, epochs=5000)
```

```python
plt.plot(hist.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.show()
```

##### 사용법은 sklearn과 비슷하다.

```python
model.predict(np.array([100, 44]).reshape(1,2))
```

```python
model.predict(np.array([60, 25]).reshape(1,2))
```

##### 가중치과 bias를 알고 싶다면

```python
W_, b_ = model.get_weights()
print('Weight is : ', W_)
print('bias is : ', b_)
```

##### 모델이 잘 만들어졌나?

```python
x = np.linspace(20, 100, 50).reshape(50,1)
y = np.linspace(10, 70, 50).reshape(50,1)

X = np.concatenate((x, y), axis=1)
Z = np.matmul(X, W_) + b_
```

```python
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs)
ax.scatter(x, y, Z)
ax.set_xlabel('Weight')
ax.set_ylabel('Age')
ax.set_zlabel('Blood fat')
ax.view_init(15, 15)
plt.show()
```

## XOR 문제
- 선형 문제로는 XOR 문제를 풀 수 없다.

```python
import numpy as np

X = np.array([
    [0, 0],
    [1, 0],
    [0, 1],
    [1, 1]
])
y = np.array([[0], [1], [1], [0]])
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(2, activation = 'sigmoid', input_shape=(2,)),
    tf.keras.layers.Dense(1, activation = 'sigmoid')
])
```

```python
model.compile(optimizer=tf.keras.optimizers.SGD(lr=0.1), loss='mse')
```

```python
model.summary()
```

- epochs : 지정된 횟수만큼 학습
- batch_size : 한번의 학습에 사용될 데이터의 수 지정

```python
hist = model.fit(X, y, epochs=5000, batch_size=1)
```

##### 학습 결과

```python
model.predict(X)
```

##### loss 상황

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(hist.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.show()
```

##### 학습에서 찾은 가중치

```python
for w in model.weights:
    print('============')
    print(w)
```

## iris 데이터

```python
from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data
y = iris.target
```

```python
y
```

##### sklearn의 one hot encoding

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(sparse=False, handle_unknown='ignore')
enc.fit(y.reshape(len(y), 1))
```

```python
enc.categories_
```

```python
y_onehot = enc.transform(y.reshape(len(y), 1))
y_onehot
```

##### 데이터 나누기

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_onehot, test_size=0.2, random_state=13)
```

### model


```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(32, input_shape=(4, ), activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])
```

##### Adam

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
hist = model.fit(X_train, y_train, epochs=100)
```

##### test 데이터에 대한 accuracy

```python
model.evaluate(X_test, y_test, verbose=2)
```

##### loss와 acc의 변화

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['accuracy'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.show()
```
