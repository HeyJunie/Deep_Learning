```python
import numpy as np
import pandas as pd

raw_link = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/x09.txt'
raw_data = np.genfromtxt(raw_link, skip_header=36)
raw_data
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline

xs = np.array(raw_data[:,2], dtype=np.float32)
ys = np.array(raw_data[:,3], dtype=np.float32)
zs = np.array(raw_data[:,4], dtype=np.float32)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs)
ax.set_xlabel('Weight')
ax.set_ylabel('Age')
ax.set_zlabel('Blood fat')
ax.view_init(15, 15)
plt.show()
```

##### 학습 대상 데이터를 추린다.

```python
x_data = np.array(raw_data[:,2:4], dtype=np.float32)
y_data = np.array(raw_data[:,4], dtype=np.float32)
y_data = y_data.reshape((25, 1))
```

```python
# pip install tensorflow
```

```python
import tensorflow as tf
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(1, input_shape=(2, ))
])
model.compile(optimizer = 'rmsprop', loss='mse')
```

```python
model.summary()
```

```python
hist = model.fit(x_data, y_data, epochs=5000)
```

```python
plt.plot(hist.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.show()
```

##### 사용법은 sklearn과 비슷하다.

```python
model.predict(np.array([100, 44]).reshape(1,2))
```

```python
model.predict(np.array([60, 25]).reshape(1,2))
```
